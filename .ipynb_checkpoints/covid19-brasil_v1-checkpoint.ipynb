***

In this notebook, I'll be analyzing Sars-CoV-2 (also know as Coronavirus or COVID-19) in Brazil. I'll be exploring:
* National data, 
* Regional data,
* State data.

The main dataset for this notebook comes from [Brazil.IO](https://brasil.io/dataset/covid19/caso_full/). Consider visiting their platform and contribute with the project.

## Preparations
***

In this part I'll be doing some preparations needed to run this dataset, like standard importations, creating some macros and other settings.

```python
### Importations

import pandas as pd
import matplotlib.pyplot as plt
import datetime
import random
import csv
import gzip
import io
from urllib.request import Request, urlopen

# To show plots properly in this notebook

%matplotlib inline
```

```python
### Creating some macros

## Date macros

today = datetime.date.today().strftime("%Y-%m-%d")
yesterday = (datetime.date.today() - datetime.timedelta(days = 1)).strftime('%Y-%m-%d')
today = yesterday
date_list = pd.date_range(start = "2020-02-25", end = today)
 
## Creating regions lists with its respective states

region_list = ["Norte", "Nordeste", "Centro-Oeste", "Sudeste", "Sul"]
sudeste_list = ["SP", "ES", "RJ", "MG"]
nordeste_list = ["AL", "BA", "PB", "PE", "SE", "PI", "CE", "MA", "RN"]
norte_list = ["AC", "AP", "AM", "TO", "PA", "RR", "RO"]
centrooeste_list = ["MT", "GO", "MS", "DF"]
sul_list = ["SC", "RS", "PR"]

# Color Palette 
color_palette = ["#54DEFD", "#F39237", "#EE4266", "#03C959", "#345995", "#5E2BFF", "#C04CFD", "#FC6DAB", "#5DD9C1", "#157F1F", "#D35F4A", "#F1C40F", "#6699CC", "#870058", "#18F2B2", "#F8333C", "#FCAB10", "#2B9EB3", "#44AF69", "#613DC1", "#0099CC", "#5B2A86", "#A5E6BA", "#D81E5B", "#FF00FF", "#E5101E", "#9E90A2"]
```

## Data Preprocessing
***

In this section I'll be downloading and cleasing the dataset. It must be said that, in order to explore data completely I'll be splitting the dataset into minor and more specific ones, such as:
* National dataset
* Regional datasets (one dataset for each region)
* State datasets (possibly one dataset containing all the states)

To get reading easier I'll go in the order said before.

### Getting the dataset ready

I'll be downloading and cleasing the dataset in order to start splitting it

```python
### Getting the dataset

def download_brasilio_table():
    request = Request("https://data.brasil.io/dataset/covid19/caso_full.csv.gz", headers={"User-Agent": "python-urllib"})
    response = urlopen(request)
    return gzip.decompress(response.read()).decode("utf-8")

covid_df = pd.DataFrame.from_dict(csv.DictReader(io.StringIO(download_brasilio_table())))
covid_df.to_csv("data/covid19-dataset-brasil-io.csv", index=False)
covid_df = pd.read_csv("data/covid19-dataset-brasil-io.csv")
```

```python
### Removing all the rows related to cities

covid_df = covid_df[~covid_df["place_type"].isin(["city"])]
```

```python
covid_df
```

```python
### Removing unnecessary columns and viewing the head of the dataset

covid_df = covid_df.drop(["city", "city_ibge_code", "estimated_population_2019", "is_last", "is_repeated",
                          "last_available_confirmed_per_100k_inhabitants",
                          "last_available_death_rate", "order_for_place", "place_type"
                         ], axis=1)
```

```python
### Reseting the indexes

covid_df.reset_index(drop=True, inplace=True)
```

```python
### Creating a list of states based on the values inside "state" column

state_list = list(dict.fromkeys(covid_df["state"].tolist()))
```

```python
### Adding a new column (REGION):

def addRegion(state):
    if state in sudeste_list:
        return "Sudeste"
    elif state in nordeste_list:
        return "Nordeste"
    elif state in norte_list:
        return "Norte"
    elif state in centrooeste_list:
        return "Centro-Oeste"
    elif state in sul_list:
        return "Sul"
    else:
        return "non identified"
    
covid_df['region'] = list(map(addRegion, covid_df.state.tolist()))
```

```python
### Changing the order of the columns

covid_df = covid_df[['date', 'last_available_date', 'region', 'state', 'epidemiological_week', 'last_available_confirmed', 'last_available_deaths', 'new_confirmed', 'new_deaths']]
```

```python
### Renaming columns

covid_df.rename(columns={"last_available_confirmed": "accumulated_cases",
                         "new_confirmed": "new_cases",
                         "last_available_deaths": "accumulated_deaths",
                         "new_deaths": "new_deaths",
                        }, inplace = True)
```

```python
### Saving the dataset

covid_df.to_csv("data/covid19-dataset-brasil-io.csv", index=False)
```

### National Dataset

In this section I'll be creating the national dataset.

```python
### Creating national data

## Function to return the data related to the country

def getNationalData():
    acc_cases = []
    acc_deaths = []
    n_cases = []
    n_deaths = []
    epi_week = []
    
    for date in date_list:
        epi_week.append(covid_df[covid_df["date"] == date.strftime("%Y-%m-%d")].epidemiological_week.iat[0])
        acc_cases.append(covid_df[covid_df["date"] == date.strftime("%Y-%m-%d")].accumulated_cases.sum())
        acc_deaths.append(covid_df[covid_df["date"] == date.strftime("%Y-%m-%d")].accumulated_deaths.sum())
        n_cases.append(covid_df[covid_df["date"] == date.strftime("%Y-%m-%d")].new_cases.sum())
        n_deaths.append(covid_df[covid_df["date"] == date.strftime("%Y-%m-%d")].new_deaths.sum())
    
    return (epi_week, acc_cases, acc_deaths, n_cases, n_deaths)

national_data = getNationalData()
```

```python
## Creating the National DataFrame

national_df = pd.DataFrame({"date": date_list,
                            "country": "Brazil", 
                            "epidemiological_week": national_data[0], 
                            "accumulated_num_cases": national_data[1],
                            "accumulated_num_deaths": national_data[2],
                            "new_num_cases": national_data[3],
                            "new_num_deaths": national_data[4]
                           })

# Deleting the variable from memory

del(national_data)
```

```python
## Saving the DataFrame

national_df.to_csv("data/covid19-brasil.csv", index=False)
```

### Regional Dataset

In this section I'll be creating the regional dataset.

```python
### Creating regional data

## Function to return the data related to the Brazil's regions

def getRegionalData():
    r_data = []
    
    for index in range(0, len(region_list)):
        acc_cases = []
        acc_deaths = []
        n_cases = []
        n_deaths = []
        temp_dataset = covid_df[(covid_df["region"] == region_list[index])]
        
        for date in date_list:
            acc_cases.append(temp_dataset[temp_dataset["date"] == date.strftime("%Y-%m-%d")].accumulated_cases.sum())
            acc_deaths.append(temp_dataset[temp_dataset["date"] == date.strftime("%Y-%m-%d")].accumulated_deaths.sum())
            n_cases.append(temp_dataset[temp_dataset["date"] == date.strftime("%Y-%m-%d")].new_cases.sum())
            n_deaths.append(temp_dataset[temp_dataset["date"] == date.strftime("%Y-%m-%d")].new_deaths.sum())
        
        r_data.append([acc_cases, acc_deaths, n_cases, n_deaths])
        
    return r_data

regional_data = getRegionalData()
```

```python
## Creating the a DataFrame for each region

norte_df = pd.DataFrame({"date": date_list,
                        "region": "Norte", 
                        "accumulated_num_cases": regional_data[0][0],
                        "accumulated_num_deaths": regional_data[0][1],
                        "new_num_cases": regional_data[0][2],
                        "new_num_deaths": regional_data[0][3]
                       })

nordeste_df = pd.DataFrame({"date": date_list,
                            "region": "Nordeste", 
                            "accumulated_num_cases": regional_data[1][0],
                            "accumulated_num_deaths": regional_data[1][1],
                            "new_num_cases": regional_data[1][2],
                            "new_num_deaths": regional_data[1][3]
                           })

centrooeste_df = pd.DataFrame({"date": date_list,
                               "region": "Centro-Oeste", 
                               "accumulated_num_cases": regional_data[2][0],
                               "accumulated_num_deaths": regional_data[2][1],
                               "new_num_cases": regional_data[2][2],
                               "new_num_deaths": regional_data[2][3]
                              })

sudeste_df = pd.DataFrame({"date": date_list,
                           "region": "Sudeste", 
                           "accumulated_num_cases": regional_data[3][0],
                           "accumulated_num_deaths": regional_data[3][1],
                           "new_num_cases": regional_data[3][2],
                           "new_num_deaths": regional_data[3][3]
                          })

sul_df = pd.DataFrame({"date": date_list,
                        "region": "Sul", 
                        "accumulated_num_cases": regional_data[4][0],
                        "accumulated_num_deaths": regional_data[4][1],
                        "new_num_cases": regional_data[4][2],
                        "new_num_deaths": regional_data[4][3]
                       })

## Creating the Regional DataFrame

regional_df = pd.concat([norte_df, nordeste_df, centrooeste_df, sudeste_df, sul_df])

# Deleting the variable from memory

del(regional_data)
```

```python
## Saving the DataFrames

norte_df.to_csv("data/covid19-norte.csv", index=False)
nordeste_df.to_csv("data/covid19-nordeste.csv", index=False)
centrooeste_df.to_csv("data/covid19-centrooeste.csv", index=False)
sudeste_df.to_csv("data/covid19-sudeste.csv", index=False)
sul_df.to_csv("data/covid19-sul.csv", index=False)
regional_df.to_csv("data/covid19-regional.csv", index=False)
```
